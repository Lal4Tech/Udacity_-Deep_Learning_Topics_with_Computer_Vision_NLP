```python
import sagemaker
from sagemaker.pytorch import PyTorch
from sagemaker.session import Session
from sagemaker import get_execution_role
```

```python
role = get_execution_role()
session = sagemaker.Session()
region = session.boto_region_name
bucket = session.default_bucket()
```

```python
output_path = f's3://{bucket}/jobs'
hyperparameters = {"epochs": "2", "batch-size": "32", "test-batch-size": "100", "lr": "0.001"}
```

```python
print(f'region: {region}')
print(f'bucket: {bucket}')
```

```python
job_name = "sagemaker-script-mode"
estimator = PyTorch(
    entry_point="scripts/pytorch_cifar.py",
    base_job_name=job_name,
    role=get_execution_role(),
    instance_count=1,
    instance_type="ml.p2.xlarge",
    use_spot_instances = True,
    checkpoint_s3_uri = f'{output_path}/{job_name}/checkpoints',
    max_run=600,
    max_wait=1200,
    hyperparameters=hyperparameters,
    framework_version="1.8",
    py_version="py36",
)
```

```python
estimator.fit(wait=True)
```
